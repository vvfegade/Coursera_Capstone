# Import libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Scrap data from wikipedia

```python
url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(url,'html.parser')
```

```python
table = soup.find('tbody')
rows = table.select('tr')
row = [r.get_text() for r in rows]
```

# creating dataframe in asked format

```python
dataframe = pd.DataFrame(row)
dataframe1 = dataframe[0].str.split('\n', expand=True)
datafeame2 = dataframe1.rename(columns=dataframe1.iloc[0])
dataframe3 = datafeame2.drop(datafeame2.index[0])
dataframe4 = dataframe3[dataframe3.Borough != 'Not assigned']
dataframe5 = dataframe4.groupby(['Postcode', 'Borough'], sort = False).agg(','.join)
dataframe5.reset_index(inplace = True)
dataframe6 = dataframe5.replace("Not assigned", "Queen's Park")
```

```python
dataframe6.head()
```

# CSV file

```python
link = "http://cocl.us/Geospatial_data"
dataframe7 = pd.read_csv(link)
dataframe7.head()
```

# Rename Postalcode as Postcode to perform merge on same attribute

```python
dataframe7.rename(columns={'Postal Code': 'Postcode'}, inplace=True)
dataframe7.head()
```

# merge

```python
dataframe8=pd.merge(dataframe6,dataframe7,on='Postcode')
```

```python
dataframe8
```

# Borough and Neighbourhood in the dataframe

```python
print('The dataframe has {} Borough and {} Neighbourhood.'.format(len(dataframe8['Borough'].unique()),dataframe8.shape[0]))
```

# creating new dataframe which Borough contains Toronto only

```python
Toronto=dataframe8[dataframe8['Borough'].str.contains('Toronto')]
```

# Generate map to show clustering of neighbouhoods

```python
address = 'Toronto'
geolocator = Nominatim(user_agent="Toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('latitude',latitude)
print('longitude',longitude)
```

```python
import folium
```

```python
MapToronto = folium.Map(location=[latitude, longitude], zoom_start=10)
```

```python
for lat, lng, borough, neighborhood in zip(Toronto['Latitude'], Toronto['Longitude'], 
                                           Toronto['Borough'], Toronto['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(MapToronto)  
    
MapToronto
```
